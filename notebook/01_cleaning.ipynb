```python
import pandas as pd

df = pd.read_csv(r"D:\My_Project\Analyse\AI_Based_Business_Analyzer\data\raw\sample_sales.csv", encoding='latin1')
df.head()

```

```python
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df =df.drop_duplicates()
```

```python
for col in ['Quantity', 'Unit Price', 'Cost', 'Discount']:
    if col in df.columns:
        df[col].fillna(0, inplace=True)
```

```python
df['Unit Price'] = df['Sales'] / df['Quantity']
df['Revenue'] = df['Quantity'] * df['Unit Price'] * (1 - df.get('Discount', 0))
df['Profit'] = df['Revenue'] - (df.get('Cost', 0) * df['Quantity'])
```

```python
df.to_csv("D:\My_Project\Analyse\AI_Based_Business_Analyzer\data\processed\cleaned_sales.csv",index=False)
print("Data cleaned and saved!")
```
